```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
# from google.colab.patches import cv2_imshow

IMG = "emma.jpg"
img = cv2.imread(IMG, cv2.IMREAD_GRAYSCALE)
if img is None:
    raise FileNotFoundError(IMG)

control_pts = [(0, 0), (49, 49),(50,100), (100, 185), (150, 255), (151, 150) , (200,200) ,(255,255)]
xp, yp = zip(*control_pts)

x = np.arange(256)
lut = np.interp(x, xp, yp).astype(np.uint8)

out = cv2.LUT(img, lut)

plt.figure()
plt.title("Q1 â€“ Intensity Transformation Function")
plt.xlabel("Input intensity")
plt.ylabel("Output intensity")
plt.plot(x, lut, color="blue")
plt.grid(True)
plt.show()

cv2.imshow("Original Image", img)
cv2.imshow("Transformed Image", out)
cv2.waitKey(0)
cv2.destroyAllWindows()


cv2.imwrite("emma_transformed.jpg", out)
```
