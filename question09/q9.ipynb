```python
import cv2, numpy as np, matplotlib.pyplot as plt

img = cv2.imread("rice.png", 0)
blur = cv2.GaussianBlur(img,(5,5),0)
_,th = cv2.threshold(blur,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
th = cv2.morphologyEx(th,cv2.MORPH_OPEN,np.ones((3,3),np.uint8),iterations=2)

n, lbl = cv2.connectedComponents(th)
print("Grains:", n-1)

plt.subplot(131);plt.imshow(img,cmap="gray");plt.title("Original");plt.axis("off")
plt.subplot(132);plt.imshow(th,cmap="gray");plt.title("Binary");plt.axis("off")
plt.subplot(133);plt.imshow(lbl,cmap="nipy_spectral");plt.title(f"Count={n-1}");plt.axis("off")
plt.show()
```
